```python
from random import uniform
n=int(input("nombre de répétition de la génération ="))
c=0
for i in range(n):
    u=uniform(0,1)
    print(u)
    if u<0.5:
        c=c+1
print("le nombre de fois où la génération est strictement plus petite que 0,5 =",c)
print("la fréquence où la génératione est plus petite que 0,5 =", c/n)
```

```python
from random import randint
a = randint(1,101)
guess=int(input("guess the number="))
while guess!=a:
    if guess > a:
        print("Your guess is too high. Guess again.")
        guess=int(input("guess the number="))
    elif guess <a:
        print("Your guess is too low. Guess again.")
        guess=int(input("guess the number="))
    else:
        pass
print("Congratulations! You guessed the number correctly!")
```
