```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def function(site):
    

    link = site
    page = requests.get(link)


    soup = BeautifulSoup(page.content,"html.parser")



    s=soup.find("span",{"class":"mw-page-title-main"}).text


    
    n= soup.find("table",{"class":"infobox biography biota"})
    v=1
    if n==None:
   
        n= soup.find("table",{"class":"infobox vevent"})
        v = 2
    tr = n.find_all('tr')
    d = {'keys':[],"values":[]}
    for t in tr[v:]:
        if t.find("th") and t.find("td"):
            d["keys"].append(t.find("th").text)
            d["values"].append(t.find("td").text)

    df = pd.DataFrame(d)



    p=soup.find_all("a")
    links = []
    for i in p:
        try:
            link = i["href"]
            if link[0]!="#":
                links.append(link)
        except:
            continue
    return df,links

function("https://en.wikipedia.org/wiki/Python_(programming_language)")
```
